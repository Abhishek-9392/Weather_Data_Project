## Summary_ Christiano_Ronaldo

- This dataset contains all the stats of all club goals of Cristiano Ronaldo dos Santos Aveiro.
- Cristiano Ronaldo dos Santos Aveiro is a Portuguese professional footballer who plays as a forward for Premier League club Manchester United and captains the Portugal national team.





## Summary of Columns:
- ***Goal_no:*** Column Describe no of Goals done by ronaldo

- ***Season:*** Year of Season 02/03 means 2002 and 2003.
- ***Competition:*** Name of the Competion under which the matches was held 
- ***Matchday:*** Describe the Praticular type or number of the match like 1st round, 2,3,4,5....etc
- ***Date:*** Particular Date of the match 
- ***Venue:*** Column describe in which stadium the match was played H=>Hotspur Stadium
- ***Club:*** Club from which ronaldo played
- ***Opponent:*** Opponent Team names
- ***Result:*** Final Score of the match 
- ***Playing_Position:*** Ronaldo playing position for particluar matched CF=> center forward
    - CF=> Center Forward
    - RW=> Right Winger
    - LW=> Left Winger
- ***Minute:*** How much time passed since match started 
- ***At_score:*** Score at the time ronaldo scored
- ***Type_of_goal:*** How he scored the goal 
- ***Goal_assist:*** Player who assisted ronaldo while scoring the goal

## Issues with Data Set:
- Matchday column has some values in number & some in round name `Consistency`

## Data Cleaning:


```python
ronaldo
```

```python
ronaldo["Goal_assist"]=ronaldo["Goal_assist"].fillna("Self")
```

```python
def replace_values(grp):
    grp["Type"]= grp["Type"].fillna(grp["Type"].mode()[0])
    grp["Playing_Position"]= grp["Playing_Position"].fillna(grp["Playing_Position"].mode()[0])

    return grp
```

```python
ronaldo=ronaldo.groupby("Season").apply(replace_values)
```

```python
ronaldo["Playing_Position"].isna().sum()
```

```python
ronaldo["Club_Result"]=ronaldo["Result"].str.split(":").str.get(0)
ronaldo["Opponent_Result"]=ronaldo["Result"].str.split(":").str.get(1).str[0]

```

```python
ronaldo["Club_At_score"]=ronaldo["At_score"].str.split(":").str.get(0)
ronaldo["Opponent_At_score"]=ronaldo["At_score"].str.split(":").str.get(1)
```

```python
ronaldo["Season"]=ronaldo["Season"].replace("Dec-13","12/13")
```

```python
ronaldo["Matchday"]=ronaldo["Matchday"].replace({
    "3rd round":"Third Round",
    "Round of 16":"16",
    "6th round":"Sixth Round",
    "last 16":"16",
    "4th round":"Fourth Round",
    "final":"Final"
})
```

```python
ronaldo["Minute"] = ronaldo["Minute"].str.split("+").apply(lambda e: int(e[0]) + int(e[1]) if len(e)>1 else int(e[0]))
```

```python
ronaldo["Date"]=pd.to_datetime(ronaldo["Date"])
```
