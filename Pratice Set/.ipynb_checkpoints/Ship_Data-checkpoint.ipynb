## Summary of the Data

### Built: 
#### This refers to the year in which the ship was constructed or built. For example, in the provided data, both ships were built in the year 2023.

### GT (Gross Tonnage): 
#### Gross Tonnage is a measure of the total internal volume of a ship, including all enclosed spaces. It is used to determine various regulatory and operational aspects, such as registration fees, safety regulations, and manning requirements. It does not directly represent the weight of the ship or cargo capacity.

### DWT (Deadweight Tonnage): 
#### The deadweight Tonnage is the total weight a ship can safely carry, including cargo, fuel, crew, provisions, etc. It represents the maximum load-carrying capacity of the vessel and is used to calculate freight rates and efficiency.

### Length: 
#### Refers length of the ship

### Width: 
#### Refers width of the ship

### Ship_name:
#### Refers the Ship name from the particular tables.

### Company_name:
#### refers the company name for which the ship is transporting goods

## Column Types
### Numerical - built_year, gt, dwt, length, width
### Categorical - ship_name
### Mixed - Company_name, size, country, port_name, locode
